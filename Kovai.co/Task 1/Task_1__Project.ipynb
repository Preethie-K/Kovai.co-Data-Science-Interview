```python
# Ratings Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=data, order=data['rating'].value_counts().index)
plt.title('Ratings Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Duration Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['duration'], bins=30, kde=True)
plt.title('Duration Distribution')
plt.xlabel('Duration (minutes)')
plt.ylabel('Count')
plt.show()
```

```python
# Top 10 Countries with the Most Content
top_countries = data['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with the Most Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
# Distribution of content by type and rating
plt.figure(figsize=(12, 6))
sns.countplot(x='rating', hue='type', data=data, order=data['rating'].value_counts().index)
plt.title('Distribution of Content by Type and Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.show()
```

**Cleaning the data and Visualizing the data!!!**

```python
# Assuming 'data' is your DataFrame
unique_ratings = data['rating'].unique()

# Print the unique ratings
print(unique_ratings)
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load your dataset
data = pd.read_csv("prime.csv")

# Data Cleaning
# Handling missing values
data.dropna(subset=["rating"], inplace=True)  # Remove rows with missing ratings

# Define the range of valid content ratings
valid_ratings = ['PG-13', 'TV-MA', 'PG', 'TV-14', 'TV-PG', 'TV-Y', 'TV-Y7', 'R', 'TV-G', 'G', 'NC-17', 'NR', 'TV-Y7-FV', 'UR']

# Filter the dataset to keep only the rows with valid content ratings
data = data[data['rating'].isin(valid_ratings)]

# Convert the "duration" column to numeric (remove 'min' and handle non-numeric values)
data['duration'] = data['duration'].str.replace(' min', '')
data['duration'] = data['duration'].str.replace(' Seasons', '')


# Handle non-numeric durations
def handle_duration(duration):
    try:
        return float(duration)
    except ValueError:
        if 'Season' in duration:
            return 450  # Assuming 1 season is approximately 450 minutes
        else:
            return np.nan

data['duration'] = data['duration'].apply(handle_duration)

# Remove rows with missing or non-convertible values in the 'duration' column
data.dropna(subset=['duration'], inplace=True)

# Continue with data cleaning and visualization steps as needed.
```

```python
# Handle any additional data cleaning steps as required
# For example, you can remove duplicate entries
data.drop_duplicates(inplace=True)

# Check for and handle any other data quality issues specific to your dataset
# For example, you can handle missing values in other columns, standardize date formats, or correct data inconsistencies.

# Now, the data is cleaned and ready for visualization.
```

**Visualizing the Cleaned dataset**

```python
# Import necessary libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Examples of visualizations you can create include:
# - Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data["rating"], bins=30, kde=True)
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# - Duration Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data["duration"], bins=30, kde=True)
plt.title("Duration Distribution of Movies and TV Shows")
plt.xlabel("Duration (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
# - Top 10 Countries with the Most Content
top_countries = data["country"].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")
plt.title("Top 10 Countries with the Most Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()

# - Content Types (Movies vs. TV Shows)
plt.figure(figsize=(8, 6))
sns.countplot(x="type", data=data)
plt.title("Count of Movies vs. TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
# - Relationship between Duration and IMDb Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x="duration", y="rating", data=data, alpha=0.5)
plt.title("Relationship between Duration and Rating")
plt.xlabel("Duration (minutes)")
plt.ylabel("Rating")
plt.show()
```
