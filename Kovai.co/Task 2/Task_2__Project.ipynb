```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Video Games Sales.csv')

# Sort the dataset by global sales in descending order
top_selling_games = data.sort_values(by='Global', ascending=False)
print(top_selling_games)

```

```python
# Select the top N games as hits (you can define N as needed)
top_n_hits = top_selling_games.head(10)  # Here, we're selecting the top 10 hits
top_n_hits
```

```python
# Display the top N hits
print("Top Hits Across the Globe:")
print(top_n_hits[['Game Title', 'Global']])
```

**Other things that can be done on the dataset**

```python
# Regional sales trends
regional_sales = data[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
print("Overall Regional Sales Trends:")
print(regional_sales)
```

```python
# Publisher performance
publisher_sales = data.groupby('Publisher')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
print("Publisher Performance:")
print(publisher_sales)
```

```python
# Yearly trends
yearly_trends = data.groupby('Year')[['Global']].sum()
print("Yearly Sales Trends:")
print(yearly_trends)
```

```python
# Correlation between critic scores and global sales
correlation = data['Global'].corr(data['Review'])
print(f"Correlation between Critic Scores and Global Sales: {correlation:.2f}")
```

```python
# Market share analysis
total_sales = data[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
market_share = (regional_sales / total_sales) * 100
print("Market Share Analysis:")
print(market_share)
```

**Some Visualizations for the dataset**

```python
#Global Sales Distribution
import matplotlib.pyplot as plt
import seaborn as sns

# Global Sales Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Global'], kde=True)
plt.title('Global Sales Distribution')
plt.xlabel('Global Sales (in millions)')
plt.show()
```

```python
#Genre wise sales
# Genre-wise Sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Genre', y='Global', data=data, estimator=sum)
plt.title('Genre-wise Global Sales')
plt.xticks(rotation=45)
plt.xlabel('Genre')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
#Platform wise sales
# Platform-wise Sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Platform', y='Global', data=data, estimator=sum)
plt.title('Platform-wise Global Sales')
plt.xticks(rotation=45)
plt.xlabel('Platform')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
#Yearly trends

yearly_trends = data.groupby('Year')['Global'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Global', data=yearly_trends)
plt.title('Yearly Game Sales Trends')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Text Analysis for Sentiment Analysis


import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download VADER lexicon if not already downloaded
nltk.download('vader_lexicon')

# Initialize the Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()

# Sample user reviews (replace with your actual reviews)
user_reviews = [
    "This game is amazing! I love it.",
    "Terrible game. I regret buying it.",
    "It's an okay game, not great but not bad either."
]

# Analyze sentiment for each review
for review in user_reviews:
    sentiment_scores = sia.polarity_scores(review)
    sentiment = 'Positive' if sentiment_scores['compound'] > 0 else 'Negative' if sentiment_scores['compound'] < 0 else 'Neutral'
    print(f"Review: {review}")
    print(f"Sentiment: {sentiment}")
```

```python
#Time Series Analysis

#create a line plot showing how global game sales have evolved over the years.
import matplotlib.pyplot as plt

# Assuming 'data' contains your dataset
yearly_sales = data.groupby('Year')['Global'].sum()

# Plotting the global sales trend over time
plt.figure(figsize=(10, 6))
yearly_sales.plot(kind='line', marker='o')
plt.title('Global Game Sales Over Time')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.show()
```
